# 6) Transformations (Q56–75)

```python
# Create a new column Age_Group (Young, Adult, Senior) 
#📝 Note: 
#Age 5–25 → Young 
#Age 26–50 → Adult 
#Age Above 50 → Senior 

if 'Age' in df.columns:
    def age_group(x):
        if x<=25: return 'Young'
        elif x<=50: return 'Adult'
        else: return 'Senior'
    df['Age_Group'] = df['Age'].apply(age_group)
```

```python
# Create a column Risk_Level based on BMI 
#📝 Note: 
#BMI < 18.5 → Low 
#BMI 18.5–25 → Normal 
#BMI 25–30 → Medium 
#BMI > 30 → High 

if 'BMI' in df.columns:
    def risk(b):
        if b<18.5: return 'Low'
        elif b<25: return 'Normal'
        elif b<30: return 'Medium'
        else: return 'High'
    df['Risk_Level'] = df['BMI'].apply(risk)
```

```python
#Convert Registration_Date to datetime 
#📝 Note: Required before extracting year or month. 


if 'Registration_Date' in df.columns:
    df['Reg_Year'] = df['Registration_Date'].dt.year
    df['Reg_Month'] = df['Registration_Date'].dt.month
```

```python
#Create column Is_Senior (Age > 60) 
#📝 Note: True if Age > 60, else False. 

if 'Age' in df.columns:
    df['Is_Senior'] = df['Age']>60
```

```python
#Create column High_Risk_Patient 
#📝 Note: 
#BMI > 30 AND
#Chronic_Condition ≠ "None" 

if 'BMI' in df.columns and 'Chronic_Condition' in df.columns:
    df['High_Risk'] = (df['BMI']>30) & (df['Chronic_Condition']!='None')
```

```python
#Convert Gender column to category type 
#📝 Note: Improves memory efficiency. 

if 'Gender' in df.columns:
    df['Gender'] = df['Gender'].astype('category')
if 'Blood_Group' in df.columns:
    df['Blood_Group'] = df['Blood_Group'].astype('category')
```

```python
#Rename column Phone_Number to Contact_Number 
#📝 Note: Use rename operation. 

if 'Phone_Number' in df.columns:
    df.rename(columns={'Phone_Number':'Contact_Number'}, inplace=True)
```

```python
#Drop Email column 
#📝 Note: Remove unnecessary or sensitive data. 

if 'Email' in df.columns:
    df.drop(columns=['Email'], inplace=True)
```

```python
# Insurance status
if 'Insurance_Provider' in df.columns:
    df['Insurance_Status'] = np.where(df['Insurance_Provider']!='Unknown','Yes','No')
```

```python
# Create column Patient_Location 
#📝 Note: Combine City and State with separator. 


if 'City' in df.columns and 'State' in df.columns:
    df['Patient_Location'] = df['City'] + ', ' + df['State']
```

```python
#Sort patients by BMI descending 
#📝 Note: Highest BMI first. 

#Reset index after sorting 
#📝 Note: Clean index after rearranging rows. 

#Select top 10 highest BMI patients 
#📝 Note: Used for high-risk analysis. 
    

if 'BMI' in df.columns:
    df = df.sort_values(by='BMI', ascending=False)
    df = df.reset_index(drop=True)
    df.head(10)
```

## 7) Advanced (Q76–90)

```python
#Find top 5 cities with highest average BMI 
#📝 Note: Group → Mean → Sort → Top 5. 

if 'City' in df.columns and 'BMI' in df.columns:
    df.groupby('City')['BMI'].mean().sort_values(ascending=False).head()
```

```python
#Find bottom 5 cities with lowest average BMI 
#📝 Note: Ascending order.
    
#Find patients at high cardiovascular risk 
#📝 Note: 
#Age > 50 
#BMI > 30 
#Chronic condition present 

if 'Age' in df.columns and 'BMI' in df.columns and 'Chronic_Condition' in df.columns:
    high_risk = df[(df['Age']>50)&(df['BMI']>30)&(df['Chronic_Condition']!='None')]
    high_risk.head()
```

```python
#Find doctor handling maximum chronic patients 
#📝 Note: Filter chronic patients, then group by doctor. 

if 'Doctor_Name' in df.columns and 'Chronic_Condition' in df.columns:
    df[df['Chronic_Condition']!='None'].groupby('Doctor_Name').size().sort_values(ascending=False)
```

```python
#Find insurance provider covering most senior citizens 
#📝 Note: Senior = Age > 60. 

if 'Age' in df.columns and 'Insurance_Provider' in df.columns:
    df[df['Age']>60].groupby('Insurance_Provider').size().sort_values(ascending=False)
```

```python
#Find patients eligible for lifestyle counseling 
#📝 Note: BMI > 25 OR Smoking = Yes OR Alcohol = Occasionally. 

if 'Smoking_Status' in df.columns and 'Alcohol_Consumption' in df.columns and 'BMI' in df.columns:
    lifestyle = df[(df['BMI']>25)|(df['Smoking_Status']=='Yes')|(df['Alcohol_Consumption']=='Occasionally')]
    lifestyle.head()
```

```python
#Identify potential ICU-risk patients 
#📝 Note: Age > 65 AND BMI > 30 AND chronic condition present. 

if 'Age' in df.columns and 'BMI' in df.columns and 'Chronic_Condition' in df.columns:
    icu = df[(df['Age']>65)&(df['BMI']>30)&(df['Chronic_Condition']!='None')]
    icu.head()
```

```python
#Find correlation between Age and BMI 
#📝 Note: Measures relationship strength. 

if 'Age' in df.columns and 'BMI' in df.columns:
    print(df[['Age','BMI']].corr())
```
